```python
#now we have a seperate dtaframe to work with
ece=[]
dec=[]
dcs=[]
cse=[]
cce=[]
mme=[]
countec=0
countde=0
countdc=0
countcc=0
countcs=0
countme=0
sumec=0
sumcs=0
sumcc=0
sumde=0
sumdc=0
summe=0
for i in range(0,470):
    Branch_name=Branch_df.loc[i,"Roll_No"]
    if(Branch_name[2:5]=="UCC"):
        countcc=countcc+1
        sumcc=sumcc+Branch_df.loc[i,"Total_Agg"]
    if(Branch_name[2:5]=="UCS"):
        countcs=countcs+1
        sumcs=sumcs+Branch_df.loc[i,"Total_Agg"]
    if(Branch_name[2:5]=="UEC"):
        countec=countec+1
        sumec=sumec+Branch_df.loc[i,"Total_Agg"]
    if(Branch_name[2:5]=="UME"):
        countme=countme+1
        summe=summe+Branch_df.loc[i,"Total_Agg"]
    if(Branch_name[2:5]=="DEC"):
        countde=countde+1
        sumde=sumde+Branch_df.loc[i,"Total_Agg"]
    if(Branch_name[2:5]=="DCS"):
        countdc=countdc+1
        sumdc=sumdc+Branch_df.loc[i,"Total_Agg"]
ece.append(countec)
ece.append(sumec)
cce.append(countcc)
cce.append(sumcc)
cse.append(countcs)
cse.append(sumcs)
dcs.append(countdc)
dcs.append(sumdc)
dec.append(countde)
dec.append(sumde)
mme.append(countme)
mme.append(summe)
```

```python
print(ece[0])
print(ece[1])
```

```python
#Agg. marks 
ece[1]=ece[1]//ece[0]
cce[1]=cce[1]//cce[0]
dec[1]=dec[1]//dec[0]
dcs[1]=dcs[1]//dcs[0]
cse[1]=cse[1]//cse[0]
mme[1]=mme[1]//mme[0]
branch_names=["CSE","ECE","MME","CCE","DEC","DCS"]
branch_wise_marks=[cse[1],ece[1],mme[1],cce[1],dec[1],dcs[1]]
print(branch_names)
print(branch_wise_marks)
```

```python
#performace of different branch students
#bar graph for better analysis
plt.bar(branch_names,branch_wise_marks)
plt.xlabel("Branch Names",fontsize=12)
plt.ylabel("Average Marks",fontsize=12)
plt.xticks(fontsize=12,rotation=30)
plt.yticks(fontsize=12)
plt.title("Brach-Wise Distribution of average marks")
```

```python
dataset_X=dataset['Lab'].values.reshape(-1,1)
dataset_X
```

```python
dataset_Y=dataset['theory_Agg'].values.reshape(-1,1)
dataset_Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(dataset_X,dataset_Y,test_size=0.20)
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model_pred=model.predict(X_test)
```

```python
#cofficients
print("Cofficients: \n", model.coef_)
```

```python
#mean squared error
print("Mean_Squared_Error:%2f"%mean_squared_error(y_test,model_pred))
model.score(X_train,y_train)
```
